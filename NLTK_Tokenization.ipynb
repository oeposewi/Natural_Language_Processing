```python
#download Dracula
dracula = requests.get('https://www.gutenberg.org/files/345/345-0.txt').content.decode('utf-8')
print(type(dracula))
print(len(dracula))
dracula[:120]
```

```python
#Tokenize your book and print how many tokens it contains
d_tokens = nltk.word_tokenize(dracula)
print(len(d_tokens))
```

```python
#Calculate and display the average number of characters per token

len(dracula)/len(d_tokens)
```

```python
#Remove the stop words and punctuation; lowercase all the remaining tokens

stop_words = set(stopwords.words('english'))

def filter_tokens(tokens):
    
    return[i.lower() for i in tokens if i.lower() not in stop_words if i.isalnum()]

filtered_d_tkns = filter_tokens(d_tokens)

filtered_d_tkns
```

```python
#Recalculate and display the total number of tokens
print(len(filtered_d_tkns))
```

```python
#Obtain a list of the number of unique tokens and print the length of this list

vocab = set(filtered_d_tkns)
vocab
```

```python
#Calculate the average frequency of appearance of each unique token
from nltk import FreqDist

freqs = FreqDist(filtered_d_tkns)

print('Average frequency of all unique tokens: ', sum(freqs.values())/len(freqs))
print('Frequencies of all tokens: ')
freqs
```

```python
#Find and display the 10 most frequently occurring tokens

freqs.most_common(n=10)
```

```python
#Count the occurrences of the token 'I' and display this as a percentage of the total number of tokens

#since I is a stopword, using unfiltered token set
freq_all = FreqDist(d_tokens)
freq_all['I']/len(d_tokens)*100
```

```python
#Find all tokens with a length (in characters) greater than five and an occurrence frequency greater than 100

for k,v in dict(freqs).items():
    if len(k) > 5:
        if v > 100:
            print(k,v)
```

```python
#Create and display a list of the 20 most frequent bigrams
from nltk.util import ngrams
d_bigrams = [' '.join(i) for i in ngrams(filtered_d_tkns, 2)]

fdist_bigrams = FreqDist(d_bigrams)
fdist_bigrams.most_common(n=20)
```

```python
#PART 2
#Go back to the “raw” text of your book and use slicing to grab one complete sentence and assign it to a string variable

sentence = dracula[2950:3086]
print(sentence)
```

```python
#Use the .findall method to split the string on white space
import re

re.findall(r'\s', sentence)
```

```python
#Replace every white-space character in the string with a dash (“-“) character
new_sent = re.sub(r'\s', '-', sentence)
new_sent
```
